# Aesthetic Data Visualization with Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading and Exploring Data

```python
# Load the built-in dataset 'tips'
tips = sns.load_dataset('tips')

# Display the first few rows
print(tips.head())
```

## Scatter Plot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='sex', size='size')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

## Line Plot

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex', style='sex', markers=True)
plt.title('Line Plot of Total Bill vs Tip')
plt.show()
```

## Line Plot with Confidence Intervals 

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Create a lineplot with standard deviation error bars
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', errorbar='sd', color='blue')

# Add a title and labels
plt.title('Line Plot with Confidence Intervals')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Show the plot
plt.show()
```

## Bar Plot

```python
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', errorbar='sd')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

## Radial Heatmaps (Using Polar Coordinates)

```python
import numpy as np

# Generate sample data
data = tips.groupby('day', observed=False)['total_bill'].mean().reset_index()
theta = np.linspace(0, 2 * np.pi, len(data), endpoint=False)

# Create a radial heatmap
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
bars = ax.bar(theta, data['total_bill'], color=sns.color_palette('husl', len(data)), alpha=0.8, width=0.5)

# Add labels
ax.set_xticks(theta)
ax.set_xticklabels(data['day'])

# Rotate the labels for better visibility
for label, angle in zip(ax.get_xticklabels(), theta):
    label.set_rotation(np.degrees(angle) - 90)
    label.set_verticalalignment('center')

plt.title('Radial Heatmap: Average Total Bill by Day', va='top')
plt.tight_layout()
plt.show()
```

## Simple Radar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Define data
categories = ['Math', 'Reading', 'Writing', 'Science', 'History']
values = [80, 70, 75, 85, 90]

# Create radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
values += values[:1]  # Repeat the first value to close the circle
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(['20', '40', '60', '80', '100'])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title('Simple Radar Chart')
plt.show()
```

## Multiple Overlapping Radar Charts

```python
# Define data
categories = ['Speed', 'Reliability', 'Comfort', 'Safety', 'Efficiency']
group_data = {
    'Model A': [70, 80, 75, 85, 90],
    'Model B': [60, 70, 80, 75, 65],
    'Model C': [80, 85, 70, 80, 90],
}

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
for model, values in group_data.items():
    values += values[:1]  # Close the circle
    ax.fill(angles, values, alpha=0.1, label=model)
    ax.plot(angles, values, linewidth=2)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.title('Overlapping Radar Charts')
plt.show()
```

## 3D Radar Chart

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 40, size=(4, len(categories)))

# Compute angles for radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# Close the circle for the radar chart
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot each group
for i, value in enumerate(values):
    data = np.concatenate((value, [value[0]]))  # Close the circle for values
    ax.plot(angles, data, zs=i, label=f'Group {i+1}', linewidth=2)

# Set axis ticks and labels
ax.set_xticks(angles[:-1])  # Exclude the last tick for proper labeling
ax.set_xticklabels(categories)  # Match the labels to the number of ticks
ax.set_yticks([10, 20, 30, 40])
ax.set_zticks(range(len(values)))
ax.set_zticklabels([f'Group {i+1}' for i in range(len(values))])

# Add legend and title
ax.legend(loc='upper left', bbox_to_anchor=(1.1, 1.1))
ax.set_title('3D Radar Chart')
plt.show()
```

# Grouped Radar Chart with Transparent Areas

```python
# Define data
categories = ['Skill A', 'Skill B', 'Skill C', 'Skill D', 'Skill E']
values_1 = [75, 80, 65, 70, 85]
values_2 = [65, 70, 75, 80, 60]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})

values_1 += values_1[:1]
values_2 += values_2[:1]

ax.fill(angles, values_1, color='green', alpha=0.25, label='Group 1')
ax.fill(angles, values_2, color='red', alpha=0.25, label='Group 2')

ax.plot(angles, values_1, color='green', linewidth=2)
ax.plot(angles, values_2, color='red', linewidth=2)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right')
plt.title('Grouped Radar Chart')
plt.show()
```
